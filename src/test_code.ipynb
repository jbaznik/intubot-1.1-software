```python


        # setup threads and processes
        videohandler = VideoHandler(
            video_stream=video_stream_1, frame_w=FRAME_W,
            frame_h=FRAME_H, start_time=start_time, recording=recording, recording_path=PATH_TO_RECORDINGS)

        
        serialhandler = SerialHandler(start_time, recording=recording, recording_path=PATH_TO_RECORDINGS)
        computervision = ComputerVision()
        driver = Driver(PATH_TO_CONTROL_VALUES, FRAME_W, FRAME_H)

        videohandler.start()  # continuously captures frames
        serialhandler.start()  # continuously read serial inputs
        computervision.load(
            model_manikin['name'],
            model_manikin['path_to_weights'],
            model_manikin['path_to_labels'],
            model_manikin['path_to_thresholds'])  # initiates PyTorch

        # set up secondary camera
        # secondary camera is only for display, it won't be used by computer vision and visual servioing.
        if video_stream_2 is not None and video_stream_2 != video_stream_1:
            videohandler_secondary = VideoHandler(
                video_stream=video_stream_2, frame_w=FRAME_W, frame_h=FRAME_H, start_time=start_time,
                recording=recording, recording_path=PATH_TO_RECORDINGS)
            videohandler_secondary.start()
        else:
            videohandler_secondary = None

    def run(self):
        """
        waits till a frame is ready.
        once a frame is ready, performs object detection on it, calculates motor controls and drives
        """
        while True:
            if not videohandler.frame_ready:  # there is no new frame -> wait or stop driving
                if not videohandler.timestamp_lastframe or time.time() - videohandler.timestamp_lastframe < TIMEOUT_TIME:
                    # TODO: why sleep here?
                    time.sleep(0.01)  # wait till a frame is ready or time out
                    continue
                else:
                    driver.emergency_stop()
            else:  # new frame is ready
                serialhandler.frame_number = videohandler.frame_number

                # freeze passed frame
                frame = videohandler.frame
                videohandler.frame_ready = False

                # perform detection
                scene = computervision.object_detection(frame)
                new_scene = True  # flag for GUI

                # loop through scene to know what has been detected
                driver.track_anatomy_detection(scene)

                # update button states
                # add logic here if auto mode should stay active when button is released
                if serialhandler.button_auto:
                    auto_mode = True
                else:
                    auto_mode = False

                # run control function
                if auto_mode and driver.auto_mode_available:
                    driver.visual_servoing(scene, serialhandler.joystick,
                                                serialhandler.analog_rocker, serialhandler.motor)
                else:
                    driver.manual_driving(serialhandler.joystick,
                                               serialhandler.analog_rocker, serialhandler.motor)

            # send command
            if driver.auto_mode_available:
                led_auto = 100
            else:
                led_auto = 0

            serialhandler.send(motors=driver.motor_pwm_command,
                                    led_auto=led_auto,
                                    led_display=100,
                                    led_laryngoscope=100,
                                    stop=driver.stop)

    def close_event(self):
        videohandler.close()
        serialhandler.close()
        running = False


if __name__ == '__main__':
    app = MainApp()
    try:
        app.run()
    except KeyboardInterrupt:
        app.close_event()
```
